### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
# ciclo 1)
#for classico

l1 = []
for n in range(5):
    if (n%2 == 0):
        l1.append(n)

print(l1)

#versione con list comprehension

l1 = [n for n in range(5) if n%2 == 0]

print("l1 con comprehension:", l1)

print("- - - - - - - - - - - - -")

l1.clear()

# ciclo 2)

check = lambda a: (a**2 - (a-1)**2) > 2


for numero in range(10):
    if check(numero):
        l1.append(numero**2)

print(l1)

#versione con list comprehension

l1 = [n**2 for n in range(10) if check(n)]
print("l1 con comprehension:", l1)


print("- - - - - - - - - - - - -")


l1.clear()

my_dict = {"a":[1,6,3],"b":[9,1,2],"c":[2,5,1]}

for key in my_dict:
    for value in my_dict[key]:
        if value%2 == 0:
            l1.append(value)

print(l1)

l1= [value for sublist in my_dict for value in my_dict[sublist] if value%2 == 0]

print("l1 con comprehension:", l1)


```
